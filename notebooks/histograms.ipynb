```python
# Imports
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Load trace and compute deltas and encodings
enc = OrdinalEncoder()
trace = np.fromfile('/traces/canneal/roitrace.bin', dtype=np.int64)
deltas = trace[1:] - trace[:-1]
trace_enc = enc.fit_transform(trace)
deltas_enc = enc.fit_transform(deltas)
```

```python
# Plot
fig, axs = plt.subplots(ncols=2, sharey=True, figsize=(14, 8))
fig.suptitle('Canneal')

fig.subplots_adjust(hspace=0.5, left=0.07, right=0.93)
ax = axs[0]
hist = ax.hist(trace, bins=100)
ax.set_xlabel('Addresses')

ax = axs[1]
hist = ax.hist(deltas, bins=100)
ax.set_xlabel('Deltas')
```
