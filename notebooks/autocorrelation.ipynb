```python
# Imports and defs
import matplotlib.pyplot as plt
import numpy as np

def plot_autocorr(a, title):
    markerline, stemline, baseline = plt.stem(a[:,1])
    plt.setp(stemline, linewidth = 1)
    plt.setp(markerline, markersize = 4)
    plt.title(title)
    plt.ylabel('Autocorrelation Coefficient')
    plt.xlabel('Lag')
    plt.grid(True)
    plt.show()
```

```python
%matplotlib inline
# Bodytrack
bodytrack = np.vstack(([0,1], np.load('/traces/bodytrack/autocorr.npy')))
plot_autocorr(bodytrack, 'Bodytrack')
```

```python
%matplotlib inline
# Canneal
canneal = np.vstack(([0,1], np.load('/traces/canneal/autocorr.npy')))
plot_autocorr(canneal, 'Canneal')
```
