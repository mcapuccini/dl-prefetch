```python
# Imports and helpers
import matplotlib.pyplot as plt
import numpy as np

def plot_autocor(a, title):
    markerline, stemline, baseline = plt.stem(a[:,1])
    plt.setp(stemline, linewidth = 1)
    plt.setp(markerline, markersize = 4)
    plt.title(title)
    plt.ylabel('Autocorrelation Coefficient')
    plt.xlabel('Lag')
    plt.grid(True)
    plt.show()
```

```python
%matplotlib inline
autocor = np.vstack(([0,1], np.load('/traces/canneal/autocor_train.npy')))
plot_autocor(autocor, 'Canneal train')
```

```python
%matplotlib inline
autocor = np.vstack(([0,1], np.load('/traces/canneal/autocor_dev.npy')))
plot_autocor(autocor, 'Canneal dev')
```

```python
%matplotlib inline
autocor = np.vstack(([0,1], np.load('/traces/canneal/autocor_test.npy')))
plot_autocor(autocor, 'Canneal test')
```
